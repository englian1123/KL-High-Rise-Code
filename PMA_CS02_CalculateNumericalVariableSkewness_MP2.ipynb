# Check the numerical variables skewness

```python
import pandas as pd
from scipy.stats import skew

# Assuming 'Transaction Price' is a column in your DataFrame 'df'
# If not, replace 'Transaction Price' with the actual column name.

# Load the data (replace with your actual file path if different)
df = pd.read_csv('https://raw.githubusercontent.com/englian1123/KL-High-Rise-Data/refs/heads/main/KLHighRise.csv')

# Calculate skewness
skewness = skew(df['TransactionPrice'])  # Correct column name
print(f"The skewness of Transaction Price is: {skewness}")

skewness = skew(df['ParcelArea'])  # Correct column name
print(f"The skewness of Land/Parcel Area is: {skewness}")
```

```python
from scipy.stats import kurtosis

# Calculate kurtosis
kurtosis_price = kurtosis(df['TransactionPrice'])
print(f"The kurtosis of Transaction Price is: {kurtosis_price}")

kurtosis_area = kurtosis(df['ParcelArea'])
print(f"The kurtosis of Land/Parcel Area is: {kurtosis_area}")
```
