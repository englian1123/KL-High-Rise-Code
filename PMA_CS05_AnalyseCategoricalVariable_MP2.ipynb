```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from IPython.display import display # Import display

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/englian1123/KL-High-Rise-Data/refs/heads/main/KLHighRise.csv")

# Create output directory
output_dir = "categorical_univariate_analysis"
os.makedirs(output_dir, exist_ok=True)

# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Perform univariate analysis for non-constant categorical variables
for col in categorical_cols:
    unique_values = df[col].nunique()
    if unique_values > 1:
        # Frequency count
        freq_table = df[col].value_counts(dropna=False).reset_index()
        freq_table.columns = [col, 'Count']
        freq_table['Percentage'] = (freq_table['Count'] / freq_table['Count'].sum()) * 100

        # Save frequency table
        freq_table.to_csv(f"{output_dir}/{col}_summary.csv", index=False)

        # Display frequency table
        print(f"\nFrequency table for {col}:") # Add a header for clarity
        display(freq_table)

        # Create bar plot using matplotlib and seaborn
        plt.figure(figsize=(10, 6)) # Adjust figure size as needed
        sns.barplot(x=col, y='Count', data=freq_table)
        plt.title(f"Frequency of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=45, ha='right') # Rotate labels if they overlap
        plt.tight_layout() # Adjust layout to prevent labels overlapping
        plt.savefig(f"{output_dir}/{col}_barplot.png")
        # plt.close() # Close the plot to free up memory - Removed to display chart

print("Univariate analysis completed for non-constant categorical variables. Outputs saved in 'categorical_univariate_analysis' folder.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import display

df = pd.read_csv('https://raw.githubusercontent.com/englian1123/KL-High-Rise-Data/refs/heads/main/KLHighRise.csv')  # load CSV

# Calculate frequency counts for categorical variables
categorical_cols = ['UnitLevel', 'Mukim', 'Tenure', 'SchemeName']
freq_counts = {}

for col in categorical_cols:
    freq_counts[col] = df[col].value_counts()

# Display frequency counts in a tabular format
for col, counts in freq_counts.items():
    print(f"Frequency counts for {col}:")
    display(counts)

# Plot settings for improved visualization
plt.rcParams.update({'font.size': 12, 'figure.figsize': (10, 6)})  # Adjust font size and figure size


# Plot horizontal bar plots for each categorical variable
for col, counts in freq_counts.items():
    plt.figure()
    display (counts)
    counts = counts.sort_values(ascending=True)
    counts.plot(kind='barh', color='skyblue') # Added color to bars
    plt.title(f"Distribution of {col}", fontsize=14)
    plt.xlabel("Frequency", fontsize=12)
    plt.ylabel(col, fontsize=12)
    plt.tight_layout()
    plt.show()

# Plot pie charts for each categorical variable
for col, counts in freq_counts.items():
    plt.figure()  # Create a new figure for each pie chart
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 10}) # Adjust text size
    plt.title(f"Distribution of {col}", fontsize=14)
    plt.axis('equal')
    plt.tight_layout()  # Adjust layout to prevent labels from overlapping
    plt.show()

# Boxplot for Transaction Price by "Tenure" with improved visualization
plt.figure(figsize=(16, 8))  # Adjust figure size for better readability
sns.boxplot(x='Tenure', y='TransactionPrice', data=df,
            order=df.groupby('Tenure')['TransactionPrice'].median().sort_values().index,
            palette="Set3")  # Use a different color palette

plt.xticks(rotation=45, ha="right", fontsize=12)  # Rotate x-axis labels and adjust font size
plt.title('Transaction Price by Tenure', fontsize=16)
plt.xlabel('Tenure', fontsize=14)
plt.ylabel('Transaction Price', fontsize=14)

plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/englian1123/KL-High-Rise-Data/refs/heads/main/KLHighRise.csv"
df = pd.read_csv(url)

# Clean columns
df['Parcel_sq_m'] = df['ParcelArea'].astype(float)  # Corrected column name and removed regex for simplicity
df['Year'] = pd.to_datetime(df['TransactionDate'], format='%b-%y').dt.year

# Additional stats
print("Parcel Area Summary:")
print(df['Parcel_sq_m'].describe())
print("\nUnitLevel Summary:")
print(df['UnitLevel'].describe())
print("\nYearly Counts:")
print(df['Year'].value_counts().sort_index())
print("\nAverage Price by Mukim:")
print(df.groupby('Mukim')['TransactionPrice'].mean().sort_values(ascending=False))
print("\nAverage Price by Tenure:")
print(df.groupby('Tenure')['TransactionPrice'].mean())

# Quick plot: Price distribution
df['TransactionPrice'].hist(bins=50, figsize=(10,5))
plt.title('Transaction Price Distribution')
plt.xlabel('Price (RM)')
plt.ylabel('Frequency')
plt.show()

# Additional stats
print("Parcel Area Summary:")
print(df['Parcel_sq_m'].describe())
print("\nUnitLevel Summary:")
print(df['UnitLevel'].describe())
print("\nYearly Counts:")
print(df['Year'].value_counts().sort_index())
print("\nAverage Price by Mukim:")
print(df.groupby('Mukim')['TransactionPrice'].mean().sort_values(ascending=False))
print("\nAverage Price by Tenure:")
print(df.groupby('Tenure')['TransactionPrice'].mean())

# Quick plot: Price distribution
df['TransactionPrice'].hist(bins=50, figsize=(10,5))
plt.title('Transaction Price Distribution')
plt.xlabel('Price (RM)')
plt.ylabel('Frequency')
plt.show()
```
